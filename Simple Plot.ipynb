```python
import math
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def Decay(N0, l, t):
    N = N0*math.exp(-l*t)
    return N

x = np.linspace(0, 10000, 1000)
y = []

for i in x:
    y.append(Decay(1, 0.005134424, i))

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title('Simple Decay')
plt.xlabel('Time (s)')
plt.ylabel('Relative Number')
```

```python
#def lamda(t_half):
#    l = math.ln(2)/t_half
#    return l

l = [4.92e-11, 1.22e-1, 9.95e+2, 3.65e-1, 7.03e+1, 3.97e+5, 1.56e+8, 5.73e+2, 5.97e+3, 1.18e+05, 0]

#for i in (t_half):
#    l_all.append(lamda(i))

m=len(l)

#product lambda
def Decay(m, l, N0, t):
    prod_final = 1
    sum_final = 0
    prod_l = 1
    for i in range (m-1):
        prod_l = prod_l*l[i]
        #print(prod_l)

    for j in range(m):
        for i in range (m):
            #product
            if i!=j:
                prod = l[i]-l[j]
                prod_final = prod*prod_final
        sum_final+=math.exp(-l[j]*t)/prod_final
        prod_final=1
    N=N0*prod_l*sum_final
    return N
```
